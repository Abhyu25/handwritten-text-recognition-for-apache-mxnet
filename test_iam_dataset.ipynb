```python
import numpy as np
import random
import time
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from IPython import display

from utils.iam_dataset import IAMDataset
```

```python
ds = IAMDataset("word", output_data="text")

fig, ax = plt.subplots(1)

for i in range(25):
    n = int(random.random()*len(ds))
    image_i, text_i = ds[n]
    plt.imshow(image_i, cmap='Greys_r')        
    plt.title(text_i[0])
    display.display(plt.gcf())
    display.clear_output(wait=True)
    time.sleep(1)
```

```python
ds = IAMDataset("form", output_data="bb", output_parse_method="line")
fig, ax = plt.subplots(1)

for i in range(25):
    n = int(random.random()*len(ds))
    image_i, bb_i = ds[n]
    plt.imshow(image_i, cmap='Greys_r')
    rects = []
    for j in range(bb_i.shape[0]):
        bb_ij = bb_i[j]
        rect = patches.Rectangle((bb_ij[0], bb_ij[1]), bb_ij[2], bb_ij[3], fill=False, color="r")
        rects.append(rect)
        ax.add_patch(rect)
        
    display.display(plt.gcf())
    display.clear_output(wait=True)
    
    time.sleep(1)
    for rect in rects:
        rect.remove()
```
